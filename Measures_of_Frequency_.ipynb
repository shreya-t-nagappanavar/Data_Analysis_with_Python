# Frequency distribution

**Frequency distribution**: is a tabular summary of data showing the frequency(or number) of items in each of serveral non-overlapping classes.


1.   Un-grouped frequency distribution: All distinct values of the variable are mentioned and their frequencies are counted.
2.   Grouped frequency distribution: Values are divided between different intervals and then their frequencies are counted

**a) Realative Frequency**

```python
total_rooms = pd.crosstab(df['latitude'], ['total_bedrooms'])
```

```python
total_rooms['relative_frequency'] = total_rooms['total_bedrooms'] / total_rooms['total_bedrooms'].sum()
```

```python
total_rooms
```

**b) Pencentage Frequency**

```python
total_rooms['percentage_frequency'] = total_rooms['relative_frequency'] * 100
```

```python
total_rooms['percentage_frequency']
```

```python
total_rooms
```

```python
df['total_rooms'].value_counts().plot(kind = 'pie', autopct = '%1.f%%')
```
