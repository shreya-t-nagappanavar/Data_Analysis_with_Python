1) Finding ethnicity by Female

Note: use values to run code properly(to get dictnary)

```python
df = df.values
```

```python
c = 0
for i in df:
  if(i[4] == 'Hispanic' and i[5] == 'F'):
    c += 1

print(c)
```

2) Finding more cgp by acedemic student 2006

```python
c = 0
for i in df:
  if(i[7] == 'Fall 2006' and i[16] >= 2.64):
    c += 1

print(c)
```

3) with fall 2006, more than 2.6 cgpa and city.

```python
c = 0
for i in df:
  if(i[7] == 'Fall 2006' and i[16] >= 2.64 and i[17] == 'New York'):
    c += 1

print(c)
```
