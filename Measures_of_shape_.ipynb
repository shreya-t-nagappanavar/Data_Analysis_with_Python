# Skewness


*   Skewness is a statistical number that tells us if a distribution is symmetric or not.


*   A distribution is symmetric if the right side of the distribution is similar to the left side of the distribution.



Symmetric data
*   mean = median = mode

Negative Direction
*   mean<median<mode


Positive Direction
*   mode<median<mean












Pearson's Coefficient of skewness

Sk = (mean - mode)/o-

if mode is not well defined

Sk = 3(mean - mode)/o-

**2) Kurtosis**

Kurtosis is a statistical number that tells us if a distribution is taller or shorter than a normal distribution.


*   If a distribution is similar to the normal distribution , the kurtosis value is 0.
*   If kutosis is greater than 0, then it has a higher peak compared to the normal distribution.


*   If kurtosis is less than 0,then it is flatter than a normal distribution.

Formula =( (E[(X - u)]4)/o-4) - 3

4 is squre root

There are three types of distributions:

*   Mesokurtic: Medium peaked
*   Platykurtic: Flattest peak and highly dispersed.
